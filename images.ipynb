## Load using `keras.preprocessing`

A simple way to load images is to use `tf.keras.preprocessing`.

```python
# The 1./255 is to convert from uint8 to float32 in range [0,1].
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

Define some parameters for the loader:

```python
BATCH_SIZE = 32
IMG_HEIGHT = 224
IMG_WIDTH = 224
STEPS_PER_EPOCH = np.ceil(image_count/BATCH_SIZE)
print("Steps per epoch: {}".format(STEPS_PER_EPOCH))
```

```python
train_data_gen = image_generator.flow_from_directory(directory=str(data_dir),
                                                     batch_size=BATCH_SIZE,
                                                     shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                     classes = list(CLASS_NAMES))
```

Inspect a batch:

```python
def show_batch(image_batch, label_batch):
  plt.figure(figsize=(10,10))
  for n in range(25):
      ax = plt.subplot(5,5,n+1)
      plt.imshow(image_batch[n])
      plt.title(CLASS_NAMES[label_batch[n]==1][0].title())
      plt.axis('off')
```

```python
image_batch, label_batch = next(train_data_gen)
```

## Load using `tf.data`

The above `keras.preprocessing` method is convienient, but has three downsides: 

1. It's slow. See the performance section below.
1. It lacks fine-grained control.
1. It is not well integrated with the rest of TensorFlow.

To load the files as a `tf.data.Dataset` first create a dataset of the file paths:

```python
str(data_dir/'*/*')
```

```python
list_ds = tf.data.Dataset.list_files(str(data_dir/'*/*'))
print(list_ds)
```

```python
for f in list_ds.take(5):
  print(f.numpy())
```

Write a short pure-tensorflow function that converts a file path to an `(img, label)` pair:

```python
f_name = list_ds.take(1)
print(list(f_name)[0].numpy())
for n in f_name:
  parts = tf.strings.split(n, os.path.sep)
  print(parts)
  print(parts[-2])
  print(parts[-2] == CLASS_NAMES)
  print(CLASS_NAMES)
```

```python
def get_label(file_path):
  # convert the path to a list of path components
  parts = tf.strings.split(file_path, os.path.sep)
  # The second to last is the class-directory
  return parts[-2] == CLASS_NAMES
```

```python
def decode_img(img):
  # convert the compressed string to a 3D uint8 tensor
  img = tf.image.decode_jpeg(img, channels=3)
  # Use `convert_image_dtype` to convert to floats in the [0,1] range.
  img = tf.image.convert_image_dtype(img, tf.float32)
  # resize the image to the desired size.
  return tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
```

```python
def process_path(file_path):
  label = get_label(file_path)
  # load the raw data from the file as a string
  img = tf.io.read_file(file_path)
  img = decode_img(img)
  return img, label
```

Use `Dataset.map` to create a dataset of `image, label` pairs:

```python
# Set `num_parallel_calls` so multiple images are loaded/processed in parallel.
labeled_ds = list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
```

```python
for image, label in labeled_ds.take(1):
  print("Image shape: ", image.numpy().shape)
  print("Label: ", label.numpy())
```

### Basic methods for training

To train a model with this dataset you will want the data:

* To be well shuffled.
* To be batched.
* Batches to be available as soon as possible.

These features can be easily added using the `tf.data` api.

```python
ex = tf.data.Dataset.from_tensor_slices([1, 2, 3])
ex = ex.repeat()
# print(list(ex.as_numpy_iterator()))
```

```python
def prepare_for_training(ds, cache=True, shuffle_buffer_size=1000):
  # This is a small dataset, only load it once, and keep it in memory.
  # use `.cache(filename)` to cache preprocessing work for datasets that don't
  # fit in memory.
  if cache:
    if isinstance(cache, str):
      ds = ds.cache(cache)
    else:
      ds = ds.cache()

  ds = ds.shuffle(buffer_size=shuffle_buffer_size)

  # Repeat forever
  ds = ds.repeat()

  ds = ds.batch(BATCH_SIZE)

  # `prefetch` lets the dataset fetch batches in the background while the model
  # is training.
  ds = ds.prefetch(buffer_size=AUTOTUNE)

  return ds
```

```python
train_ds = prepare_for_training(labeled_ds)

image_batch, label_batch = next(iter(train_ds))
```

```python
show_batch(image_batch.numpy(), label_batch.numpy())
```

# tf.keras.utils.Sequence

```python
def get_label_conventional(file_path):
  # convert the path to a list of path components
  parts = str(file_path).split(os.path.sep)
  # The second to last is the class-directory
  return parts[-2] == CLASS_NAMES
```

```python
def decode_img_conventional(img):
  # convert the compressed string to a 3D uint8 tensor
  img = tf.keras.preprocessing.image.load_img(img, color_mode='rgb', target_size=[IMG_HEIGHT, IMG_WIDTH])
  # Use `convert_image_dtype` to convert to floats in the [0,1] range.
  img = np.array(img).astype('float')
  img = img / 255.0
  # resize the image to the desired size.
  return img
```

```python
def process_path_conventional(file_path):
  label = get_label_conventional(file_path)
  img = decode_img_conventional(str(file_path))
  return img, label
```

```python
data = list(data_dir.glob('*/*.jpg'))[:32]
image_batch = []
label_batch = []
for fp in data:
  img, lbl = process_path_conventional(fp)
  image_batch.append(img)
  label_batch.append(lbl)
show_batch(image_batch, label_batch)
```

```python
from tensorflow.python.keras.utils.data_utils import Sequence
import random
```

```python
class DataGenerator(Sequence):
    def __init__(self, x_set, batch_size, shuffle=False):
        self.x = x_set
        self.batch_size = batch_size
        self.shuffle = shuffle
        if shuffle:
          self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.x) / float(self.batch_size)))

    def __iter__(self):
      for item in (self[i] for i in range(len(self))):
        yield item

    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.x) 

    def __getitem__(self, idx):
        batch_x = self.x[idx * self.batch_size:(idx + 1) * self.batch_size]
                
        images = []
        labels = []
        for fp in batch_x:
          img, lbl = process_path_conventional(fp)
          images.append(img)
          labels.append(lbl)
            
        return images, labels
```

```python
train_data_seq = DataGenerator(list(data_dir.glob('*/*.jpg')), 32, shuffle=True)
```

```python
imgs, lbls = next(iter(train_data_seq))
show_batch(imgs, lbls)
```

## Performance

Note: This section just shows a couple of easy tricks that may help performance. For an in depth guide see [Input Pipeline Performance](../../guide/performance/datasets).

To investigate, first here's a function to check the performance of our datasets:

```python
import time
default_timeit_steps = 1000

def timeit(ds, steps=default_timeit_steps):
  start = time.time()
  it = iter(ds)
  for i in range(steps):
    try:
      batch = next(it)
    except StopIteration:
      it = iter(ds)
    if i%10 == 0:
      print('.',end='')
  print()
  end = time.time()

  duration = end-start
  print("{} batches: {} s".format(steps, duration))
  print("{:0.5f} Images/s".format(BATCH_SIZE*steps/duration))
```

Let's compare the speed of the two data generators:

```python
# `keras.preprocessing`
timeit(train_data_gen)
```

```python
# `sequence'
timeit(train_data_seq)
```

```python
# `tf.data`
timeit(train_ds)
```

A large part of the performance gain comes from the use of `.cache`.

```python
uncached_ds = prepare_for_training(labeled_ds, cache=False)
timeit(uncached_ds)
```

If the dataset doesn't fit in memory use a cache file to maintain some of the advantages:

```python
filecache_ds = prepare_for_training(labeled_ds, cache="./flowers.tfcache")
timeit(filecache_ds)
```
